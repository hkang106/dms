```python
import cv2
import numpy as np
from glob import glob
import os
from matplotlib import pyplot as plt
```

```python
IMAGE_PATH="./dataset/widerface/val/images/0--Parade/0_Parade_Parade_0_194.jpg"
```

```python
img1 = cv2.imread(IMAGE_PATH, cv2.IMREAD_COLOR)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
f1 = [111, 425, 122, 127]
f2 = [209, 347, 70, 103]
f3 = [368, 252, 89, 133]
f4 = [555, 282, 89, 100]
f5 = [707, 252, 92, 133]
```

```python

x1 = f1[0]
y1 = f1[1]
x2 = x1+f1[2]
y2 = y1+f1[3]


x11 = f2[0]
y11 = f2[1]
x22 = x11+f2[2]
y22 = y11+f2[3]


x111 = f3[0]
y111 = f3[1]
x222 = x111+f3[2]
y222 = y111+f3[3]


x1111 = f4[0]
y1111 = f4[1]
x2222 = x1111+f4[2]
y2222 = y1111+f4[3]


x11111 = f5[0]
y11111 = f5[1]
x22222 = x11111+f5[2]
y22222 = y11111+f5[3]




green_color = (0,255,0)

cv2.line(img1, (x1,y1), (x1, y2), green_color, 3)
cv2.line(img1, (x1,y2), (x2, y2), green_color, 3)
cv2.line(img1, (x2,y2), (x2, y1), green_color, 3)
cv2.line(img1, (x2,y1), (x1, y1), green_color, 3)

cv2.line(img1, (x11,y11), (x11, y22), green_color, 3)
cv2.line(img1, (x11,y22), (x22, y22), green_color, 3)
cv2.line(img1, (x22,y22), (x22, y11), green_color, 3)
cv2.line(img1, (x22,y11), (x11, y11), green_color, 3)


cv2.line(img1, (x111,y111), (x111, y222), green_color, 3)
cv2.line(img1, (x111,y222), (x222, y222), green_color, 3)
cv2.line(img1, (x222,y222), (x222, y111), green_color, 3)
cv2.line(img1, (x222,y111), (x111, y111), green_color, 3)

cv2.line(img1, (x1111,y1111), (x1111, y2222), green_color, 3)
cv2.line(img1, (x1111,y2222), (x2222, y2222), green_color, 3)
cv2.line(img1, (x2222,y2222), (x2222, y1111), green_color, 3)
cv2.line(img1, (x2222,y1111), (x1111, y1111), green_color, 3)

cv2.line(img1, (x11111,y11111), (x11111, y22222), green_color, 3)
cv2.line(img1, (x11111,y22222), (x22222, y22222), green_color, 3)
cv2.line(img1, (x22222,y22222), (x22222, y11111), green_color, 3)
cv2.line(img1, (x22222,y11111), (x11111, y11111), green_color, 3)



```

```python
plt.figure(figsize=(25,15))
plt.imshow(img1)
plt.show()
```
