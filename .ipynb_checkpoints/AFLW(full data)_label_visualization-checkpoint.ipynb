```python
img1 = cv2.imread(image1_path, cv2.IMREAD_COLOR)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
img1.shape
```

```python
for idx in range(0, len(features1), 2):
    f = (int(features1[idx]), int(features1[idx+1]))
    cv2.circle(img1, f, 2, (0,133,225, -1))
```

```python
plt.figure(figsize=(25,15))
plt.imshow(img1)
plt.show()
```
